```python
from numpy.random import seed
import numpy as np
import tensorflow
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import OneHotEncoder
```

```python
runtime = pd.DataFrame(runtime)
```

```python
df.loc[df["runtime"] <= 90, "runtime"] = 1
df.loc[df["runtime"].between(90,120), "runtime"] = 2
df.loc[df["runtime"] >= 120, "runtime"] = 3
```

```python
df
```

```python
df = df.dropna().reset_index()
df
```

```python
df = df.drop(columns=["index","Unnamed: 0"])
df
```

```python
label_encoder_production = LabelEncoder()
le_pc = label_encoder_production.fit(df["production_company"])

label_encoder_rating = LabelEncoder()
le_rating = label_encoder_rating.fit(df["content_rating"])
```

```python
oe_style = OneHotEncoder()
oe_results = oe_style.fit_transform(df["production_company"].values.reshape(-1,1))
pc_df = pd.DataFrame(oe_results.toarray(), columns=oe_style.categories_[0])
pc_df = pc_df.astype(int)
pc_df
```

```python
oe_style_2 = OneHotEncoder()
oe_results_2 = oe_style.fit_transform(df["content_rating"].values.reshape(-1,1))
cr_df = pd.DataFrame(oe_results_2.toarray(), columns=oe_style.categories_[0])
cr_df = cr_df.astype(int)
cr_df
```

```python
full_df = pd.concat([df, pc_df, cr_df], axis=1)
full_df
```

```python
X = full_df.drop(columns=["audience_rating","content_rating","production_company"])
y = full_df[['audience_rating']].values.reshape(-1,1)
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

X_train.head()
```

```python
model = Sequential()
model.add(Dense(units=1500, activation='relu', input_dim=702))
model.add(Dense(units=1500, activation='relu'))
model.add(Dense(units=1500, activation='relu'))
model.add(Dense(units=1000, activation='relu'))
model.add(Dense(units=1, activation='linear'))
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='mean_squared_error')
```

```python
model.fit(
    X_train,
    y_train,
    epochs=1000,
    shuffle=True,
    verbose=2
)
```

```python
model.predict(X_test.iloc[15].values.reshape(1,-1))
```
