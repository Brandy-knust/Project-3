```python
rot_df = pd.read_csv('rotten_tomatoes_movies_sql.csv')
```

```python
rot_df=rot_df.drop(columns=['rotten_tomatoes_link', 'critics_consensus', 'movie_info', 'authors', 'streaming_release_date'])
rot_df
```

```python
# rot_df.parse(['original_release_date']).dt
```

```python
rot_df['year']=rot_df['original_release_date'].str[-4:]
```

```python
rot_df['year']
```

```python
rot_df['new_movie_title']=rot_df['movie_title']+"("+rot_df['year']+")"
rot_df=rot_df.set_index('new_movie_title')
rot_df
```

```python
rot_df['audience_count'].fillna(value=0, inplace=True)
```

```python
rot_df.sort_values("audience_count", ascending=False)
```

```python
filtered_rot_df = rot_df[rot_df.audience_count>25000].copy()
filtered_rot_df
```

```python
filtered_rot_df=filtered_rot_df.dropna()
```

```python
genres_list = filtered_rot_df["genres"].tolist()
genres_list
```

```python
movie_genre=['new_movie_title']
for genre in genres_list:
  genre = str(genre)
  movie_genre.extend(genre.split(', '))
Genre_names = set(movie_genre)
len(Genre_names)

movies2=filtered_rot_df.genres.str.split(',', expand=True)
movies2_df = pd.DataFrame(movies2)
columns = []
movies2_df
```

```python
movie_genre_df = pd.DataFrame(columns=Genre_names)
movie_genre_df = movie_genre_df.set_index('new_movie_title')
movie_genre_df
```

```python
for i, movie in filtered_rot_df.iterrows():
  in_movie = movie['genres'].split(', ')
  data_dictionary={}

  for genre in in_movie:
    data_dictionary[genre]=1
  data_dictionary['new_movie_title']=i
  movie_genre_df = movie_genre_df.append(data_dictionary, ignore_index=True)
```

```python
#Actors_string = filtered_movies_df.loc[:,"actors"].astype(str).copy()
#filtered_movies_df.loc[:, 'actors'] = Actors_string.copy()
#Directors"] = filtered_movies_df["directors"].astype(str)
filtered_rot_df["actors"] = filtered_rot_df["actors"].astype(str)
filtered_rot_df["directors"] = filtered_rot_df["directors"].astype(str)
filtered_rot_df["genres"] = filtered_rot_df["genres"].astype(str)
```

```python
movie_genre_df = movie_genre_df.set_index('new_movie_title')
movie_genre_df
```

```python
directors_list = filtered_rot_df["directors"].tolist()
directors_list
```

```python
movie_director=['new_movie_title']
for director in directors_list:
  # director = str(director)
  Directors=director.split(',')
  Directors = [x.strip() for x in Directors]
  movie_director.extend(Directors)
Director_names = set(movie_director)
len(Director_names)
```

```python
movie_director_df = pd.DataFrame(columns=Director_names)
movie_director_df=movie_director_df.set_index('new_movie_title')
movie_director_df
```

```python
for i, movie in filtered_rot_df.iterrows():
  in_movie = movie['directors'].split(', ')
  data_dictionary={}

  for director in in_movie:
    data_dictionary[director]=1
  data_dictionary['new_movie_title']=i
  movie_director_df = movie_director_df.append(data_dictionary, ignore_index=True)
```

```python
# movie_genre_df = movie_genre_df.set_index('new_movie_title')
# movie_genre_df

movie_director_df = movie_director_df.set_index('new_movie_title')
movie_director_df
```
